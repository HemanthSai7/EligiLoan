```python
import numpy as np
import pandas as pd
```

```python
train_df=pd.read_csv("./dataset/train.csv")
test_df=pd.read_csv("./dataset/test.csv")
```

```python
# Lowercasing column names and removing whitespaces and "-".
train_df.columns = train_df.columns.str.lower().str.replace("-", "").str.replace(" ", "_").str.replace("__", "_")
test_df.columns = test_df.columns.str.lower().str.replace("-", "").str.replace(" ", "_").str.replace("__", "_")
```

```python
# Lowercasing all object elements in the dataset
train_df = train_df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
test_df = test_df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
```

```python
# Renaming Home ownership and employment duration columns as the column names seem to have been swapped
train_df.rename(columns={'home_ownership': 'employment_duration',
             "employment_duration": "home_ownership"}, inplace=True)
test_df.rename(columns={'home_ownership': 'employment_duration',
            "employment_duration": "home_ownership"}, inplace=True)
```
